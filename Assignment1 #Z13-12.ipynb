```python
import json
import requests

url = "https://raw.githubusercontent.com/mitre/cti/master/enterprise-attack/enterprise-attack.json"

response = requests.get(url)
data = response.json()

techniques = []

for obj in data["objects"]:
    if obj.get("type") == "attack-pattern":
        techniques.append(obj)

print("Total Attack Techniques Found:", len(techniques))

def threat_score(tech):
    score = 5

    name = tech.get("name", "").lower()
    description = tech.get("description", "").lower()

    if "credential" in name:
        score += 3
    if "execution" in name:
        score += 2
    if "privilege" in name:
        score += 3
    if "persistence" in name:
        score += 2
    if "lateral" in name:
        score += 2

    if "administrator" in description:
        score += 2
    if "remote" in description:
        score += 2
    if "bypass" in description:
        score += 2
    if "stealth" in description:
        score += 1

    return score

scored_techniques = []

for tech in techniques:
    score = threat_score(tech)
    scored_techniques.append((tech.get("name", "Unknown"), score))

scored_techniques.sort(key=lambda x: x[1], reverse=True)

print("\nTop 10 Highest-Risk Techniques:\n")

for name, score in scored_techniques[:10]:
    print(f"{name} — Score: {score}")

print("\nCritical Threats (Score >= 8.9):\n")

for name, score in scored_techniques:
    if score >= 8.9:
        print(f"{name} — Score: {score}")
```

This scoring system helps cybersecurity analysts prioritize threats by automatically assigning risk scores based on intelligence indicators found in technique names and descriptions. Techniques involving credentials, privilege escalation, remote access, or bypassing security controls are scored higher because they pose greater organizational risk. By sorting techniques based on score, analysts can quickly focus on the most critical threats instead of manually reviewing thousands of entries.
